```python
#cleaning.
nation = nation.replace(['-~0.0'],'0.0')
nation = nation.replace(['~0.0'],'0.0')

nation['GDP growth rate (annual %, const. 2005 prices)'] = nation['GDP growth rate (annual %, const. 2005 prices)'].astype(float)
nation['Economy: Agriculture (% of GVA)'] = nation['Economy: Agriculture (% of GVA)'].astype(float)
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(nation.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
sns.heatmap(nation.corr())
```

```python
# normalizing both the variables
gdp_mean = nation['GDP: Gross domestic product (million current US$)'].mean()
gdp_growth_rate_mean = nation['GDP growth rate (annual %, const. 2005 prices)'].mean()

gdp_std = nation['GDP: Gross domestic product (million current US$)'].std()
gdp_growth_rate_std = nation['GDP growth rate (annual %, const. 2005 prices)'].std()


nation['GDP: Gross domestic product (million current US$)'] = nation['GDP: Gross domestic product (million current US$)'].map(lambda p: (p - gdp_mean)/(gdp_std))
nation['GDP growth rate (annual %, const. 2005 prices)'] = nation['GDP growth rate (annual %, const. 2005 prices)'].map(lambda p: (p - gdp_growth_rate_mean)/(gdp_growth_rate_std))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x = nation['GDP: Gross domestic product (million current US$)'] ,y = nation['GDP growth rate (annual %, const. 2005 prices)'])
plt.xlabel('GDP')
plt.ylabel('GDP growth rate')
```

```python
nation['Economy: Agriculture (% of GVA)'] = nation['Economy: Agriculture (% of GVA)'].astype(float)
```

```python
# dropping the columns with negative percentage contribution.

nation = nation[nation['Economy: Agriculture (% of GVA)'] > 0.0] 
nation = nation[nation['Economy: Services and other activity (% of GVA)'] > 0.0]
nation = nation[nation['Economy: Industry (% of GVA)'] > 0.0]
```

```python
agri = nation.groupby('Region')['Economy: Agriculture (% of GVA)'].mean()
industry = nation.groupby('Region')['Economy: Industry (% of GVA)'].mean()
tertiary = nation.groupby('Region')['Economy: Services and other activity (% of GVA)'].mean()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()
for col in nation.columns.values[2:]:
    if ((col!='GDP: Gross domestic product (million current US$)')&(col!='Labour force participation (female/male pop. %)')):
        try:
            corr_to_gdp[col] = nation['GDP: Gross domestic product (million current US$)'].corr(nation[col])
        except Exception as e:
            print("")
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP: Gross domestic product (million current US$)', data=nation
                   ,ax=axes[i,j], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```
